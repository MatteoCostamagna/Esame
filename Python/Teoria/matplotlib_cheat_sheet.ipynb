# **PLOT**


---



```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,  #-------------> valori ascisse
         y1, #-------------> valori ordinate
         label="sin", #----> etichetta da visualizzare 
         color="blue") #---> assegna un colore 

plt.plot(x, 
         y2, 
         label="cos", 
         color="green")

plt.title("tracciati") #-------> titolo del grafico
plt.xlabel('asse x') #----> etichetta per l'asse delle x
plt.ylabel('asse y') #----> etichetta per l'asse delle y


# definisce i punti di minimo e massimo con 
# i quali iniziare e termninare il disegno 
# di un grafico.
# N.B. Utile nel disengo dei tracciati.
plt.xlim(xmin=0)

# se si prova a imporre un limite minimo
# al grafico in esempio, si noter√† come 
# il disegno viene tagliato verso il basso
# plt.ylim(ymin=0)


# il parametro bbox_to_anchor, definisce 
# la posizione del box di visualizzazione 
# della legenda.
plt.legend(bbox_to_anchor=(1.2, 1.03)) 


plt.show() # stamapa 
```

# **SCATTER**


---



```python
sx = np.array([2,3,8,9,15,18,21,22,23,24,36,41,47,54,62])
sy = np.array([85, 80, 75, 86, 84, 90, 98, 96, 92, 93, 105, 95,  94, 105, 100])

plt.scatter(sx, 
            sy, 
            color = "blue", 
            s     = 80,  # ----> definisce le dimensioni dei punti
            alpha = 0.7,  # ----> definisce un livello di trasparenza ai punti del grafico
            label = "dati")
                         


plt.title("grafico di dispersione") 
plt.xlabel('asse x') 
plt.ylabel('asse y')               
plt.show()
```

# **PIE**


---



```python
# dati
d1 = np.arange( 2, 50, 5, np.float64)
d2 = np.arange(50, 80, 5, np.float64)
d3 = np.arange(80, 120, 5, np.float64)
d4 = np.arange(120, 160, 5, np.float64)


# totali
tot_d1 = sum(d1)
tot_d2 = sum(d2)
tot_d3 = sum(d3)
tot_d4 = sum(d4)
tot    = tot_d1 + tot_d2 + tot_d3 + tot_d4


# percentuali
perc_d1 = (tot_d1 / tot) * 100.
perc_d2 = (tot_d2 / tot) * 100.
perc_d3 = (tot_d3 / tot) * 100.
perc_d4 = (tot_d4 / tot) * 100.


# visualizzazione
sezioni   = np.array([perc_d1,   perc_d2,  perc_d3,    perc_d4])
etichette = np.array(["foo",     "baar",   "baz",      "qux"])
colori    = np.array(["#789fde", "#78de97", "#ded678", "#9d8fdb"])
offsets = np.array([0.,         0.1,       0.2,       0.])
bordi = {"linewidth": 1,  "edgecolor":"#6f6a85"} 


plt.pie(sezioni,
        labels = etichette, 
        explode = offsets, #------> separa gli spicchi a cui √® stato dato un offset diverso da 0
        autopct = '%2.2f%%', #----> stampa delle percentuali 
        shadow = True, #----------> disegna un'ombreggiatura sul bordo
        wedgeprops = bordi, # definisce uno stile di separazione tra gli spicchi
        colors = colori)    

plt.show()
```

# **SUBPLOT AND STYLES**


---



```python
x = np.linspace(0, 10, 100)
y = np.cos(x)


# imposta le dimensioni per l'intero disegno
plt.figure(figsize=(8, 12))
# spaziatura verticale
plt.subplots_adjust(hspace=0.4)

# il metodo "subplot(row, col, num)" permette di gestire pi√π grafici nello stesso layout
# now = quanti grafici impilati su riga
# col = quanti grafici affiancati in colonna
# num = numero di grafici totale 

# ======
# COLORE
# ======
cl = plt.subplot(2, 1, 1)
cl.plot(x,y, color="green")
cl.set_title("colore")
cl.set(xlabel='asse x', ylabel='axx y')

# ===========
# STILE LINEA
# ===========
sl = plt.subplot(2, 1, 2)
sl.plot(x,y, color="black", linestyle="--")
sl.set_title("stile linea")
sl.set(xlabel='asse x', ylabel='axx y')

# ====================
# SPESSORE LINEA LINEA
# ====================

plt.show()
```
