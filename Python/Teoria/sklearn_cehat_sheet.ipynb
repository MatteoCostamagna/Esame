# **Composizione della libreria scikit-learn**


La libreria **sklearn** è disgnata attorno ad acluni specifici prinicipi condivisi da tutte le librerie software che a vario livello implementano modelli di Machine Learning.

- **Estimators**: Sono tutte quelle classi o moduli che applicano un modello matematico eseguendo una stima sui valori di un dataset. Ne è un esempio il metodo ***fit()***, con il quale viene attivato il processo di learning per tutti i modelli di ***ML***.

- **Transformers**: Alune classi ***Estimator*** eseguono operatzioni esplicte sul dataset trasformandone i valori. Nella libreria **sklearn** la classe ***SimpleImputer*** ad esempio, usata per completare i valori mancanti in una tabella. Una classe di questo tipo implementa sempre un metodo ***transform()***.

- **Predictors**: Sono tutte quelle classi che implementano metodi dedicati alla classificazione e alla predizione. Contengono sempre almeno due metodi: ***predict()***, per la predizione e ***score()*** che restituisce le misure relative alla qualità della predizione.

- **Inspection**: Tutte le classi ***Estimator*** hanno parametri e iperparametri come attributi della classe. In genere vengono valorizzati durante l'esecuzione di un metodo e hanno tutti per convenzione un underscore come suffisso (es. ***coef_***).

## **Cross Validation Test**

Ogni volta che si divide un dataset per ricavare il set di training e il set di validazione, si rischia di causare un alterzione dei dati.
Potrebbe capitare, infatti, di escludere dal set di training dati fondamentali per la risucita dell'apprendimento oppure di creare un set di tes poco coerente con la descrizione dei dati.

Una tecncia molto semplice per testare le performance del modello con i dati splittati secondo la nostra impostazione, viene chiamata **k-fold validation**.

Parendo da un esempio di split che vede l'80% dei dati assegnati al ***training*** e il 20% al ***test***.
Si parte con l'addetramento prendendo i blocchi di test in coda all'array di dati, si completa il giro ***traingin-test***, dopodichè si scala di volta in volta per k-volte il blocco di test verso la testa dell'array finché non saranno esaurite le combinazioni possibili.

Alla fine avremo un insieme di valori che ci diranno, per ogni iterazione,quale sia la combinazione più performante.

## **Confusion matrix**

- **TP (True Positive)**:  numero dei campioni positivi correttamente previsti
- **TN (True Negative)**:  numero dei campioni negativi correttamente previsti
- **FP (False Positive)**: numero dei negativi erroneamente previsti come positivi
- **FN (False Negative)**: numero dei positivi erroneamente previsti come netagivi


- **TPR (True Positive Rate)**: $\frac{TP}{TP + FN}$

- **TNR (True Negative Rate)**: $\frac{TN}{TN + FP}$

- **FPR (False Positive Rate)**: $\frac{FP}{FP + TN}$

- **FPR (False Negative Rate)**: $\frac{FN}{FN + TP}$

- **Accuratezza (Accuracy)**: $\frac{TP + TN}{P + N} = \frac{Previsioni \ Corrette}{Totale \ Campioni}$

- **Precisione**: $\frac{TP}{TP + FP}$
